<h3>Understanding the self-attention mechanism from scratch </h3>

The following is the formula for attention: 
$Attention(Q, K, V) = softmax(\frac{Q.K^T}{\sqrt{d_{k}}}).V$

<br>
The idea is to understand the significance of this technique & what it does from scratch by emulating the formula for simple vectors.

```python
import numpy as np
from scipy.special import softmax
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
# initialize the input vectors of 3Dimensions for Attention mechnaism
# this is a simplified demonstration of the vectorized representation of a token in the corpus
# on top of which Attention is to be used
# The objective is to assess the attention scores betwee these 3 vectors

arg_1 = np.random.randint(-1, 2, size=3) # vector for Token 1
arg_2 = np.random.randint(-1, 1, size=3) # vector for Token 2
arg_3 = np.random.randint(-2, 1, size=3) # vector for Token 3
```

```python
%matplotlib inline

fig = plt.figure()
fig.set_size_inches(18, 10)

ax = fig.add_subplot(111, projection="3d")

# starting coordinate for plotting the vectors is (0, 0, 0)
ax.quiver(0, 0, 0, *arg_1, color="red")
ax.quiver(0, 0, 0, *arg_2, color="blue")
ax.quiver(0, 0, 0, *arg_3, color="green")

ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])

plt.show()
```

```python
# Initialize the Q, K, V Matrices
# These weight matrices are learnt bu Backpropagation - just demonstrating how they're used in
# self-attention mechanisms i.e. in the forward-propagation step, for simplicity we assume that
# the Q, K, V matrices have accurate representations post-backpropagation.

# Obtain the Query Matrix
WQ = np.random.randint(0, 3, size=(3, 3))

# Obtain the Key Matrix
WK = np.random.randint(0, 3, size=(3, 3))

# Obtain the Value Matrix
WV = np.random.randint(0, 3, size=(3, 3))

# Get a preview of these matries
display("Query Matrix:", WQ, "Key Matrix:", WK, "Value Matrix:", WV)
```

```python
# Process the Input vectors of tokens with respective Q, K, V weight matrices
input_vectors = np.vstack([arg_1, arg_2, arg_3])

# preview the input vectors
display(input_vectors)
```

```python
# Obtain the Query representation of all the input vectors by taking the dot product of the vector with the Query
Qvector = input_vectors.dot(WQ) # Query Vector matrix

display(Qvector)
```

```python
# Obtain the Key representation of all the input vectors by taking the dot product of the vector with the Key
Kvector = input_vectors.dot(WK) # Key Vector matrix

display(Kvector)
```

```python
# Obtain the Value representation of all the input vectors by taking the dot product of the vector with the Value
Vvector = input_vectors.dot(WV) # Key Vector matrix

display(Vvector)
```

```python
# get the number of dimensions in the Key matrix & get the square-root i.e. sqrt(dk), we expect the Key representation
# to be a square matrix i.e. same number of rows & columns

dimensions_K = np.array( [Kvector.shape[0]], dtype="float32" )

# preview the dimension
dimensions_K
```

```python
# obtain the square-root of the dimension
dk = np.sqrt(dimensions_K)

# preview
dk
```

```python
# Calculate the Self-Attention matrix using the formula

# get the numerator of the Self-Attention mechanism
selfAttentionNumerator = Qvector.dot(Kvector.transpose())

# preview
display(selfAttentionNumerator)
```

```python
# Utilize the denominator i.e. sqrt of dimension of Key representation vector to get argument of soft-max
selfAttentionRawValue = selfAttentionNumerator / dk

# preview
display(selfAttentionRawValue, dk)
```

```python
# Apply Softmax activation 

# There are 2 axes in the selfAttention arg. matrix i.e. row-wise i.e. per token & columns-wise 
# we want to apply softmax per token i.e. per row in the 2D self-atttention matrix 
softmaxSelfAttention = softmax(selfAttentionRawValue, axis=1)

# preview
display(softmaxSelfAttention)
```

```python
# Obtain the final Attention Score by taking the dot-product with Value vector representation
selfAttentionScore = softmaxSelfAttention.dot(Vvector)

# preview
display(selfAttentionScore)
```

```python
# preview the Attention scores
ax = sns.heatmap(selfAttentionScore, linewidth=0.5, cmap="rocket")
```
